The cells below import the data from 2010-2022 and the corresponding mappings to the Development Region codes. It merges the two, cleans the numeric data by converting it to the raw number (values were stored as thousands). Then it groups by the codes as the data from 2010-2022 was stored as towns, and multiple town corresponded to the Development Regions.

```python
df1 = pd.read_csv('cleaned/tourism_data_2010-2022.csv')
maps = pd.read_csv('mappings_10-22.csv')
```

The cell below just imports the population by year and by region data.

```python
population = pd.read_csv('population_clean.csv')
```

```python
temp = pd.merge(df1, maps[['Municipality','Town']], on='Municipality', how='left')
temp = temp.iloc[1:,:]
temp = temp.sort_values('Town')
for col in temp.columns[2:-1]:
    temp[col] = temp[col].str.replace(',','').astype(float)
temp.iloc[:,2:-1] = temp.iloc[:,2:-1]*1000
part1 = temp.groupby('Town').sum().reset_index()
```

```python
part1
```

The cells below imports the data from 2000-2009 and the corresponding mappings. It cleans the mappings a bit, taking only the Development Regions, and then performs a group by on the development regions, similar to the previous part.

```python
df2 = pd.read_csv('cleaned/tourism_data_2000-2009.csv')
maps2 = pd.read_csv('cleaned/tourism_data_2000-2009.csv')
maps2 = maps2[maps2['GEO'] == 'RD'][['NUM','NAME']].drop_duplicates().reset_index(drop=True)
maps2['NAME'] = maps2['NAME'].str[:-3]
maps2['NAME'] = maps2['NAME'].str.capitalize()
```

```python
part2 = df2[df2['NUM'].str[:2] == 'RD'].groupby('NUM').sum().reset_index()
```

```python
part2
```

Next, the cells merge the two parts, combining all years. It drops the months from 2022, and then renames the columns.

```python
merged = pd.merge(part2, part1, left_on='NUM', right_on='Town', how='inner')
merged = merged.drop('Town', axis=1)
merged = merged.iloc[ :,:-7]
merged2 = pd.merge(maps2, merged, on='NUM', how='inner')
```

```python
columns = []
for i in range(2000,2022):
    for j in range(1,13):
        columns.append(str(i)+'-'+str(j))
columns = ['Code','Name'] + columns
merged2.columns = columns
```

Certain places have tourism data for two regions in one. Because I can't seperate these regions, they become one.

```python
merged2['Code'] = merged2['Code'].str[:4]
merged2['Region Number'] = merged2['Code'].str[2:].astype(int)
merged3 = merged2[['Region Number'] + list(merged2.columns)[:-1]]
```

```python
merged3
```

The data is then melted. Instead of each column being a year, data is made vertical with the date being a part of the row. The date is then split, so year and month are seperate columns. For further explanation of the function pd.melt(), see pandas documentation: https://pandas.pydata.org/docs/reference/api/pandas.melt.html.


```python
merged4 = pd.melt(merged3, id_vars=['Region Number','Code','Name'], value_vars=list(merged3.columns)[3:])
```

```python
merged4['Year'] = merged4['variable'].str[:4].astype(int)
merged4['Month'] = merged4['variable'].str[5:].astype(int)
```

```python
merged5 = merged4[['Region Number','Code', 'Name', 'Year','Month','value']]
```

```python
merged5.columns = ['Region Number','Code', 'Name', 'Year','Month','Tourism Money']
```

```python
merged5
```

Now the data is augmented with the population for the region of that year. This way the data can be tourism money per capita.

```python
population
```

```python
merged6 = pd.merge(merged5, population, on=['Region Number','Year'], how='inner').drop('Name_y', axis=1)
merged6.columns = ['Region Number', 'Code', 'Name', 'Year', 'Month', 'Tourism Money', 'Population']
merged6['Tourism PC'] = (merged6['Tourism Money'] / merged6['Population']).astype(float)
```
