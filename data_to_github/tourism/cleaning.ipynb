```python
import numpy as np
import pandas as pd
```

Data is in a shape where we have quarterly data and annual data, along with room revenue broken down by type of accomodation. Cell below takes in the sheets of the 2000-2009 data, and gets only the monthly data, and the totals for the regions.

```python
output = []
for sheet in ['2000-2003','2004-2007','2008-2009']:
    df = pd.read_excel('hand_cleaned/tourism_data_2000-2010.xlsx', sheet_name=sheet)
    temp = [x for x in df.columns if str(x)[:3] == '200' and str(x)[-1] != '1']
    cols = list(df.columns[:4]) + temp
    data = df[cols]
    temp2 = [str(x)[:10] for x in df.columns if str(x)[:3] == '200' and str(x)[-1] != '1']
    new_cols = ['GEO','NUM','TYPE','NAME'] + temp2
    data.columns = new_cols
    data2 = data.iloc[1:,:]
    data3 = data2[data2['TYPE'] == 'TOT'].reset_index(drop=True)
    output.append(data3)
```

Cell takes the three different sheets and merges them such that all years/months are on the columns, and then checks that the merge went well (No duplicated rows for regions). Then, it gets only the desired columns to produce a final table.

```python
merge_temp = pd.merge(output[0], output[1], on='NAME', how='left')
merge_final = pd.merge(merge_temp, output[2], on='NAME', how='left')
for col in ['GEO','NUM','TYPE']:
    temp = sum((merge_final[col] == merge_final[col+'_x']) & (merge_final[col] == merge_final[col+'_y']) & (merge_final[col+'_y'] == merge_final[col+'_x']))
    if (temp != 93):
        print(col)
    else:
        print('Good!')
temp_cols =[x for x in merge_final.columns if x[:3] == '200']
final_cols = ['GEO','NUM','NAME'] + temp_cols
final_data = merge_final[final_cols]
```
