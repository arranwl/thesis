```python
weather = pd.read_csv('../weather/clean_weather.csv')
fire = pd.read_csv('../fires/clean_fire_data.csv')
tourism = pd.read_csv('../tourism/clean_tourism.csv')
```

Merging all the data and remove colder months.

```python
temp = pd.merge(fire, weather, on=['Weather City','Year','Month'], how='left')
```

```python
temp2 = pd.merge(temp, tourism, on=['Region Number','Year','Month'], how='inner').dropna()
```

```python
temp2 = temp2[temp2['Month'].isin([4,5,6,7,8,9])].reset_index(drop=True)
```

```python
temp2
```

The cells below get all the desired columns, then groups by Region, Year and Month, getting the means and sums of all the columns. 

```python
temp3 = temp2[['Year', 'Month', 'Region Name', 'Region Area', 'Mean Max Temp', 'Mean Min Temp',
       'Mean Temp', 'Total Rain', 'Total Snow', 'Total Precip','Tourism PC', 'Lightning Dummy', 'SIZE_HA']]
```

```python
temp4 = temp3.groupby(['Region Name','Year','Month']).mean().reset_index()
temp_sums = temp3.groupby(['Region Name','Year','Month']).sum().reset_index()
```

```python
temp4
```

```python
temp_sums
```

The cells below reaplce the SIZE_HA from the averages with the sum, to get the total area burned in a region per year and month. A column with the percentage of area burned is also created. Then the columns are renamed, and all values in 2022 are dropped.

```python
temp4['SIZE_HA'] = temp_sums['SIZE_HA']
```

```python
temp4['Fire Area Perc'] = (temp4['SIZE_HA'] / temp4['Region Area'])*100
```

```python
temp5 = temp4.drop('Region Area', axis=1).dropna()
```

```python
temp5
```

```python
temp5.columns = ['Region Name', 'Year', 'Month', 'Mean Max Temp', 'Mean Min Temp',
       'Mean Temp', 'Total Rain', 'Total Snow', 'Total Precip',
       'Tourism PC', 'Lightning Perc','Fire Area', 'Fire Area Perc']
```

```python
temp6 = temp5[temp5['Year'] != 2022].reset_index(drop=True)
```

```python
temp6
```

A version of the final data is saved for the visualizations.

```python
temp6.to_csv('all_data_no_dums.csv', index=False)
```

Dummy variables are then created for all controls: Regions, Years, Months. The base cases, those that are droppped, are Alberni-Clayquot, 2001, and April respectively.

```python
region_dums = pd.get_dummies(temp6['Region Name']).drop('Alberni-Clayquot',axis=1)
```

```python
year_dums = pd.get_dummies(temp6['Year']).drop(2001, axis=1)
```

```python
month_dums = pd.get_dummies(temp6['Month']).drop(4,axis=1)
```

```python
final_data = pd.concat([region_dums, year_dums, month_dums, temp6.drop(['Region Name','Year','Month'],axis=1)],axis=1)
```

```python
final_data = final_data.dropna().reset_index(drop=True)
```
